```python
X = seve.drop('quality', axis=1)
y = seve['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Prsence de sulfates et d'acide citrique

```python
issa = sns.boxplot(x='quality',y='sulphates', data=seve)
issa.set(title="Sulphates in Different Quality Wines")
```

### Pourcentage d'alcool

```python
hamza = sns.boxplot(x='quality',y='alcohol', data=seve)
hamza.set(title="Alcohol Percent in Different Quality Wines")
```

### Normaliser les donnes d'entranement

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

### Crer un modle de rgression linaire et l'entraner sur les donnes d'entranement

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

###  Normaliser les donnes de test et les utiliser pour prdire la qualit du vin

```python
X_test = scaler.transform(X_test)
y_pred = model.predict(X_test)
```

### valuer les performances du modle en utilisant R2 score

```python
bray = r2_score(y_test, y_pred)
print(f"R2 score: {bray}")
```

### preduction 

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

### Crer une visualisation graphique des prdictions de qualit du vin

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=3)
ax.set_xlabel('Vraie qualit du vin')
ax.set_ylabel('Qualit du vin prdite')
plt.show()
```
